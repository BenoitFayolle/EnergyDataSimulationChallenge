```python
import pandas as pd
```

```python
import statsmodels.api as sm
```

```python
from statsmodels.tsa.arima_process import arma_generate_sample
```

```python
# 期間の設定：365日
```

```python
nobs = 365
```

```python
# 時系列インデックスの生成
```

```python
dates = pd.date_range(start='2014-12-01', periods=nobs)
```

```python
# MAパラメーターを設定：今回はARモデルなのでMA(0)とします
```

```python
import numpy as np
```

```python
maparams = np.array([1])
```

```python
# φ=1.01としてサンプルデータを生成しグラフにプロット
```

```python
arparams = np.array([1, -1.01])
```

```python
y1 = pd.TimeSeries(arma_generate_sample(arparams, maparams, nobs), index=dates)
```

```python
y1.plot()
```

```python
# φ=0.5としてサンプルデータを生成しグラフにプロット
```

```python
arparams = np.array([1, -0.5])
```

```python
y2 = pd.TimeSeries(arma_generate_sample(arparams, maparams, nobs), index=dates)
```

```python
y2.plot()
```

```python
# φ=-1.01としてサンプルデータを生成しグラフにプロット
```

```python
arparams = np.array([1, 1.01])
```

```python
y3 = pd.TimeSeries(arma_generate_sample(arparams, maparams, nobs), index=dates)
```

```python
y3.plot()
```
